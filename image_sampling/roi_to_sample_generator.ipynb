```python
import sys
import os
from PIL import Image

sys.path.append(os.path.abspath('..'))
from detection.lib.model.ImageROI import csv_to_roi
```

```python
image_dir = '/home/rahul/data/raw_data/annotated_jpegs_all'
roi_dir = '/home/rahul/data/zooniverse/raw_bounding_boxes'
out_dir = '/home/rahul/data/zooniverse/sampled_images'

print('image_dir: ', image_dir)
print('roi_dir: ', roi_dir)
print('out_dir: ', out_dir)
```

```python
roi_files = [os.path.join(roi_dir, f) for f in os.listdir(roi_dir)
            if os.path.isfile(os.path.join(roi_dir, f)) and f.endswith('.csv')]

for roi_file in roi_files:
    print('roi_file: ', roi_file)
    roi_name = os.path.basename(roi_file)
    img_name = roi_name[:roi_name.rfind('.csv')]
    img_basename, img_ext = os.path.splitext(img_name)
    img_path = os.path.join(image_dir, img_name)
    img = Image.open(img_path)
    rois = csv_to_roi(roi_file)
    
    for ind, roi in enumerate(rois):
        print(roi.toString())
        out_path = os.path.join(out_dir, str(img_basename) + "-" + str(ind) + str(img_ext))
        cropped_img = img.crop((roi.x, roi.y, roi.x + roi.width, roi.y + roi.height))
        cropped_img.convert('RGB').save(out_path)
```
