```python
import sys, os
import numpy as np
import pandas as pd
```

```python
zooniverse_filename = '/home/rahul/data/zooniverse/logs/stripped-book-annotation-classification-classifications-11June2018.csv'
print('zooniverse_filname: ', zooniverse_filename)
```

```python
all_classifications = pd.read_csv(zooniverse_filename)
print('columns: ' , all_classifications.columns)
```

```python
# annotations = all_classifications[['classification_id', 'annotations', 'subject_data']]
# print(annotations.head())
```

```python
import json
sys.path.append(os.path.abspath('..'))
from detection.lib.model.ImageROI import ImageROI, roi_to_csv

image_rois = {}

for i in range(len(all_classifications)):
    annotobj = json.loads(all_classifications.iloc[i]['annotations'])
    bboxes = [bbox
              for jo in annotobj if 'box around each annotation' in jo['task_label']
              for bbox in jo['value']]
    
    rois = [ImageROI(x = bbox['x'],
                     y = bbox['y'],
                     width = bbox['width'],
                     height = bbox['height'],
                     isAnnotation = True)
            for bbox in bboxes]
    
    subjobj = json.loads(all_classifications.iloc[i]['subject_data'])
    image_name = subjobj[list(subjobj.keys())[0]].get('filename', None)
    
    if image_name is not None and (len(rois) > 0):
        if image_name not in image_rois:
            image_rois[image_name] = []
        image_rois[image_name].extend(rois)


outdir = 'test'
if not os.path.exists(outdir):
    os.makedirs(outdir)
for image_name in image_rois:
    print(image_name, len(image_rois[image_name]))
    roi_to_csv(image_rois[image_name], os.path.join(outdir, str(image_name)+".csv"))
    
```
